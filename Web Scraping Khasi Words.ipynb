```python
import requests
import urllib.request
import time
from bs4 import BeautifulSoup
```

```python
bilingual_data = []
headers = requests.utils.default_headers()
headers['User-Agent'] = 'Mozilla/5.0 (X11; Linux x86_64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/56.0.2924.87 Safari/537.36'
r = requests.get('https://academic.oup.com/journals', headers=headers)
```

```python
for i in range(96,101):
    url = "https://tatoeba.org/eng/sentences/search?query=&from=kha&to=eng&page="+str(i)
    print("Fetching page "+ str(i))
    print("*"*20)
    response = requests.get(url,headers=headers)
    soup = BeautifulSoup(response.text, "html.parser")
    results = soup.findAll("div", attrs={'class': 'sentence-and-translations'})
    records = []
    for result in results:
        try:
            #print(result)
            kha = result.find('div', attrs={'class':'sentence'}).find('div', attrs={'class':'text'}).text # result not results
            eng = result.find('div', attrs={'class':'direct translations'}).find('div', attrs={'class':'text'}).text
        except:
            print("!!!!!!!!!!!!!!Cannot fetch page "+ str(i))
            continue
        records.append((kha, eng))
    bilingual_data.append(records)
#print(response)
```
